```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap 
```

```python
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
```

```python
death = pd.read_csv('Test4.csv')
death
```

```python
death.dtypes
```

```python
fig = plt.figure(figsize = (9,9))
ax = fig.add_subplot(111, projection = '3d')
ax.set_title('Deaths Due to Abuse of Drugs in 2020 in Germany', fontweight ='bold', fontsize = 15)

ax.set_xlabel('Due to Alcohol', fontweight ='bold')
ax.set_ylabel('Due to other Psychoactive Drugs', fontweight ='bold')
ax.set_zlabel('Total Deaths', fontweight ='bold')


x = death['Alcohol Use'] 
y = death['Drug Use'] 
z = death['Total deaths both']



trisurf = ax.plot_trisurf(x, y, z,cmap = 'GnBu', linewidth = 0.2, antialiased = True, edgecolor = 'black') 
fig.colorbar(trisurf, ax = ax, shrink = 0.5, aspect = 5)

plt.savefig('Germany_deaths20.png', dpi=250)

plt.show()
```
